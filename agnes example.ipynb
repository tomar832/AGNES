# read data/import dataset


```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
```

# read data/Importing the dataset

read the dataset " mall customers.csv" into a dataframe

```python
df = pd.read_csv('Mall_Customers.csv')
X = df.iloc[:, 3:].values
```

set attributes, only using the last two colums,specifically the colums starting from index 3 

```python
df.head(2)
```

# Get method/ Dendogram

agglomerative method,starts from a single sample and builds up a to single cluster final 

```python
import scipy.cluster.hierarchy as sch
dendogram = sch.dendrogram(sch.linkage(X, method='ward', metric='euclidean'))
plt. title('Dendogram')
plt.xlabel('customers')
plt.ylabel('Euclidean distance')# # distance between two points
```

0-200 customers

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters=5)# # (n_cluster=None, distance_threshold=100)
hc.fit(X)
```

if we set the distance treshold to 100 then we will have  clusters 

```python
hc.n_clusters_
```

# Visualising the clusters

```python
for i in range(hc.n_clusters_):
    plt.scatter(X[hc.labels_ == i, 0], X[hc.labels_ == i, 1], label='Cluster ' + str(i))
    
    plt.title('Clusters')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
```
